## Decision Tree Classifier

Iris Dataset is used for decision tree classifier

```python
# loading the libraries
from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
```

```python
# loading the data
iris = datasets.load_iris()
features = iris.data
target = iris.target
```

```python
# creating decsion tree
decisiontree = DecisionTreeClassifier(random_state=0)
```

```python
# train the model
model = decisiontree.fit(features, target)
```

```python
# making new observation
observation = [[5, 4, 3, 2]]
```

```python
# predict observation class
model.predict(observation)
```

```python
# view predicted class probabilites
model.predict_proba(observation)
```

```python
# creating decision tree classifier using entropy
decisiontree_entropy = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
# training the model
model_entropy = decisiontree_entropy.fit(features, target)
```

## Visualizing the model

```python
# loading the libraries
import pydotplus
from IPython.display import Image
from sklearn import tree
```

```python
# creating the dot data
dot_data = tree.export_graphviz(decisiontree, out_file=None, 
                               feature_names = iris.feature_names,
                               class_names=iris.target_names)
```

```python
# draw the graph
graph = pydotplus.graph_from_dot_data(dot_data)
```

```python
# viewing the graph
Image(graph.create_png())
```

## Decision Tree Regressor


Boston dataset is used

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
# loading data with only two features
boston = datasets.load_boston()
features1 = boston.data[:,0:2]
target = boston.target
```

```python
# create decision tree
decisiontree_reg = DecisionTreeRegressor(random_state=0)
```

```python
# training the model
model = decisiontree_reg.fit(features1, target)
```

```python
# making new observation
observation1 = [[0.02, 16]]
```

```python
# predict observation value
model.predict(observation1)
```

constructing the tree to reduce mean absolute error

```python
decisiontree_mae = DecisionTreeRegressor(criterion='mae', random_state=0)
```

```python
# training the model
model_mae = decisiontree_mae.fit(features1, target)
```

```python
model_mae.predict(observation1)
```
